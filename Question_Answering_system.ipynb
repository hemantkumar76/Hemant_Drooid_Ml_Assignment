```python
# solve by hemant Kumar
import json
import pandas as pd

with open('news.article.json', 'r') as file:
    data = json.load(file)

articles = pd.json_normalize(data)
articles['dateModified'] = pd.to_datetime(articles['dateModified.$date'])
articles['scrapedDate'] = pd.to_datetime(articles['scrapedDate.$date'])

articles.head()
```

```python
israel_hamas_articles = articles[articles['articleBody'].str.contains('Israel') & articles['articleBody'].str.contains('Hamas')]

israel_hamas_articles.head()
```

```python
import re
# function banate hai for removing the special characters and all 
def clean_text(text):
    text = re.sub(r'\s+', ' ', text) 
    text = re.sub(r'[^\w\s]', '', text)  
    return text

israel_hamas_articles['cleanedArticleBody'] = israel_hamas_articles['articleBody'].apply(clean_text)

israel_hamas_articles.head()
```

```python
from transformers import pipeline

qa_pipeline = pipeline('question-answering', model='bert-large-uncased-whole-word-masking-finetuned-squad')
```

```python
def answer_question(question, context):
    results = qa_pipeline(question=question, context=context, topk=3)  
    answers = [result['answer'] for result in results]
    return answers

def ask_question(question, articles_df):
    context = ' '.join(articles_df['cleanedArticleBody'].tolist())
    answers = answer_question(question, context)
    return answers

question = "what happened at the Shifa Hospital?"
answers = ask_question(question, israel_hamas_articles)
for i, answer in enumerate(answers, 1):
    print(f"Q: {question}\nA{i}: {answer}")
```
